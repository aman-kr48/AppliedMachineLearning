```python
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit
RANDOM_SEED = 100
```

```python
def load_data(path):
    df = pd.read_csv(path, 
                     delimiter='\t', 
                     names = ['label', 'message'])
    return df
    

def preprocess_data(df):
    df1 = df.copy()
    df1['label'] = df1['label'].map({'ham': 0, 'spam': 1})
    df1 = df1.drop_duplicates()

    return df1
    

def split_data(df):
    split1 = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=RANDOM_SEED)
    split2 = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=RANDOM_SEED)

    for train_index, temp_index in split1.split(df, df["label"]):
        train = df.iloc[train_index]
        temp = df.iloc[temp_index]

    for val_index, test_index in split2.split(temp, temp["label"]):
        val = temp.iloc[val_index]
        test = temp.iloc[test_index]

    return train, val, test


def save_splits(train, val, test):
    train.to_csv('../data/train.csv', index = False)
    val.to_csv('../data/validation.csv', index = False)
    test.to_csv('../data/test.csv', index = False)
```

```python
raw_data = load_data('../data/SMSSpamCollection')
raw_data.to_csv('../data/raw_data.csv')
```

```python
preprocessed_df = preprocess_data(raw_data)
train, val, test = split_data(preprocessed_df)
save_splits(train, val, test)
```
